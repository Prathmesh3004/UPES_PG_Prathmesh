">
  <div style="font-size:10px;letter-spacing:3px;text-transform:uppercase;color:#c0392b;font-weight:700;margin-bottom:6px;">&#9993; Den Den Mushi Log</div>
  <div style="font-size:14px;line-height:1.65;"><p style="margin:0 0 8px 0;">The program defines a function that finds the maximum and minimum values in a sequence without using the built-in <code style="background:rgba(0,0,0,0.35);padding:1px 5px;border-radius:4px;">max()</code> and <code style="background:rgba(0,0,0,0.35);padding:1px 5px;border-radius:4px;">min()</code> functions. It assumes the first element is both the maximum and minimum, then compares each remaining element to update the values — true shipwright craftsmanship.</p></div>
</div>

```python
def find_max_min(numbers):
    maximum = numbers[0]
    minimum = numbers[0]

    for num in numbers:
        if num > maximum:
            maximum = num
        if num < minimum:
            minimum = num

    return maximum, minimum

numbers = [25, 12, 89, 43, 7, 56]

max_value, min_value = find_max_min(numbers)

print("Sequence:", numbers)
print("Maximum =", max_value)
print("Minimum =", min_value)
```

</div>

```python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

def print_series(terms):
    for i in range(terms):
        print(fibonacci(i), end=" ")

terms = 10

print("Fibonacci Series:")
print_series(terms)
```

</div>

```python
class EmptyFileError(Exception):
    pass


class InvalidRecordError(Exception):
    pass


def read_student_marks(file_name):
    try:
        with open(file_name, "r", encoding="utf-8") as file:
            lines = [line.strip() for line in file if line.strip()]
    except FileNotFoundError as error:
        raise FileNotFoundError(f"{file_name} not found.") from error

    if not lines:
        raise EmptyFileError(f"{file_name} is empty.")

    marks_data = {}
    for line in lines:
        parts = line.split(",")
        if len(parts) != 2:
            raise InvalidRecordError(f"Invalid line format: '{line}'")

        name = parts[0].strip()
        try:
            marks = float(parts[1].strip())
        except ValueError as error:
            raise InvalidRecordError(f"Marks must be numeric in line: '{line}'") from error

        marks_data[name] = marks

    return marks_data


with open("students_valid.txt", "w", encoding="utf-8") as file:
    file.write("Aditi,88 Rohan,76")

with open("students_invalid.txt", "w", encoding="utf-8") as file:
    file.write("Aditi-88 Rohan,abc")

for candidate in ["missing.txt", "students_invalid.txt", "students_valid.txt"]:
    try:
        data = read_student_marks(candidate)
        print(f"{candidate} -> {data}")
    except (FileNotFoundError, EmptyFileError, InvalidRecordError) as error:
        print(f"{candidate} -> {error}")
```

Each zombie needs its own stats and a way to check if it "passed" Hogback's revival process. Add constructor-based initialization for student details and implement methods:

* display() for student details
* marks_percentage() for percentage of each student
* result() where the pass condition is marks > 40 in each subject Also write a function to find class average.

<div class="op-explain" style="
  position:relative; margin:6px 0 18px 0; padding:14px 18px;
  background:rgba(11,61,92,0.5); border:1px solid #1e844988;

">
  <div style="font-size:10px;letter-spacing:3px;text-transform:uppercase;color:#1e8449;font-weight:700;margin-bottom:6px;">&#9993; Den Den Mushi Log</div>
  <div style="font-size:14px;line-height:1.65;"><p style="margin:0 0 8px 0;">Methods encapsulate operations on object data — each zombie knows how to report its own stats. A separate function computes crew-level statistics from all the objects.</p></div>
</div>

```python
class StudentRecord:
    def __init__(self, name, sap_id, marks):
        self.name = name
        self.sap_id = sap_id
        self.marks = marks

    def display(self):
        print(f"Name: {self.name} | SAP ID: {self.sap_id} | Marks: {self.marks}")

    def marks_percentage(self):
        return sum(self.marks.values()) / len(self.marks)

    def result(self):
        return "Pass" if all(mark > 40 for mark in self.marks.values()) else "Fail"


def class_average(records):
    return sum(record.marks_percentage() for record in records) / len(records)


input_data = [
    ("Riya", "5001", {"phy": 78, "chem": 67, "maths": 81}),
    ("Kunal", "5002", {"phy": 88, "chem": 92, "maths": 85}),
    ("Meera", "5003", {"phy": 39, "chem": 73, "maths": 64}),
]

records = [StudentRecord(name, sap_id, marks) for name, sap_id, marks in input_data]

for record in records:
    record.display()
    print("Percentage:", f"{record.marks_percentage():.2f}%")
    print("Result:", record.result())
    print("-" * 45)

print("Class average percentage:", f"{class_average(records):.2f}%")
```

**Question 7**

Sengoku wants the entire war visualized — troop trends, damage reports, and territory share — all on one command screen. Create a program to demonstrate different visual forms using Matplotlib.

<div class="op-explain" style="
  position:relative; margin:6px 0 18px 0; padding:14px 18px;
  background:rgba(11,61,92,0.5); border:1px solid #c0392b88;
  border-left:5px solid #c0392b; border-radius:0 10px 10px 0;
  font-family:'Segoe UI','Helvetica Neue',Arial,sans-serif; color:#f5f1e6;
">
  <div style="font-size:10px;letter-spacing:3px;text-transform:uppercase;color:#c0392b;font-weight:700;margin-bottom:6px;">&#9993; Den Den Mushi Log</div>
  <div style="font-size:14px;line-height:1.65;"><p style="margin:0 0 8px 0;">This example shows line, bar, scatter, and pie charts in one figure to compare different visualization styles — a full war-room dashboard at a glance.</p></div>
</div>

```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [120, 140, 135, 160, 175, 190]
expenses = [95, 105, 102, 110, 120, 130]

categories = ["Rent", "Food", "Transport", "Others"]
share = [35, 25, 20, 20]

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

axes[0, 0].plot(months, sales, marker="o", color="teal")
axes[0, 0].set_title("Line Chart: Monthly Sales")
axes[0, 0].set_ylabel("Sales")
axes[0, 0].grid(True, linestyle="--", alpha=0.5)

axes[0, 1].bar(months, expenses, color="orange")
axes[0, 1].set_title("Bar Chart: Monthly Expenses")
axes[0, 1].set_ylabel("Expenses")

axes[1, 0].scatter(sales, expenses, color="crimson")
axes[1, 0].set_title("Scatter Plot: Sales vs Expenses")
axes[1, 0].set_xlabel("Sales")
axes[1, 0].set_ylabel("Expenses")

axes[1, 1].pie(share, labels=categories, autopct="%1.1f%%", startangle=90)
axes[1, 1].set_title("Pie Chart: Expense Distribution")

plt.tight_layout()
plt.show()
```
